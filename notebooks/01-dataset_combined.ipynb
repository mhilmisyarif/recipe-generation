```python
import json
import re
import pandas as pd
import tqdm
import glob
```

```python
#file path with glob
path = '../data/kaggle/'
all_files = glob.glob(path + "/*.csv")

mapper = map(lambda x: pd.read_csv(x), all_files)
li = list(mapper)
df = pd.concat(li)
df.drop(columns=['URL','Loves'], inplace=True)

df.columns = df.columns.str.lower()

#change columne data dype to string
list_column = ['title','ingredients','steps']
for i in list_column:
    df[i] = df[i].astype(str)
```

```python
#seperae ingredients and steps
df['ingredients'] = df['ingredients'].str.split('--')
df['steps'] = df['steps'].str.split('--')
```

```python
#fike path with glob
path1 = '../data/mr/'
all_files = glob.glob(path1 + "/*.csv")

mapper = map(lambda x: pd.read_csv(x), all_files)
li = list(mapper)
df1 = pd.concat(li)
df1.rename(columns= {'step': 'steps'}, inplace=True)
```

```python
def remove_un(text):
    text = re.sub(r'\\n', '', text)                
    text = re.sub(r'[\[\'\]\:\-]', '', text)   
    return text

def remove_title_un(text):
    text = re.sub(r'(,.*)','',text)
    text = re.sub(r'Resep','',text)
    text = re.sub(r'Untuk.*','',text)
    text = re.sub(r'Cara Membuat ','',text)
    return text

df1['title'] = df1['title'].apply(remove_un)
df1['title'] = df1['title'].apply(remove_title_un)
df1['ingredients'] = df1['ingredients'].apply(remove_un)
df1['steps'] = df1['steps'].apply(remove_un)
```

```python
df1['ingredients'] = df1['ingredients'].str.split(',')
df1['steps'] = df1['steps'].str.split(',')
df1
```

```python
df = pd.concat([df, df1])
df
```

```python
df.to_csv('../data/combined.csv', index=False)
```

```python
import tqdm

# add into list of dictionary
list_dict = []
for i in tqdm.tqdm(range(len(df))):
    dict_ = {}
    dict_['id'] = i
    dict_['title'] = df['title'].iloc[i]
    dict_['ingredients'] = df['ingredients'].iloc[i]
    dict_['steps'] = df['steps'].iloc[i]
    list_dict.append(dict_)
```

```python
with open('../data/indo_recipe.json', 'w') as f:
    json.dump(list_dict, f)
```
